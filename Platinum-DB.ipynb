```python
import pandas as pd
```

```python
import sqlite3

# Create / Connect Database

conn = sqlite3.connect('platinum1.db') # Open a database File

print('Sukses')
```

```python
# CREATE TABLE analisis sentimen (input)

query = '''CREATE TABLE analisis_sentimen (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            flag VARCHAR(255),
            sentimen VARCHAR(255),
            tweet_baru VARCHAR(255),
            tweet VARCHAR(255)
          )'''

conn.execute(query)
conn.commit()
```

```python
# CREATE TABLE upload analisis sentimen (input)

query = '''CREATE TABLE upload_analisis_sentimen (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            flag VARCHAR(255),
            sentimen VARCHAR(255),
            tweet_baru VARCHAR(255),
            tweet VARCHAR(255)
          )'''

conn.execute(query)
conn.commit()
```

```python
#CEK TABLE YG ADA DI DATABASE KITA
query = ''' SELECT name FROM sqlite_master WHERE type='table';
 '''

conn.execute(query)

conn.commit()
df = pd.read_sql_query(query, conn)
df.head(10)
```

```python
#BACA ISI DATA

query = ''' select * from analisis_sentimen '''
df = pd.read_sql_query(query, conn)
df.head(50)
```

```python
#BACA ISI DATA

query = ''' select * from upload_analisis_sentimen '''
df = pd.read_sql_query(query, conn)
df.head(50)
```
